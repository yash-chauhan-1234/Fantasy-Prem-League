```python
points=optimisation_df["Points"]
cost=optimisation_df["Cost"]
position=optimisation_df["Position"]
name=optimisation_df["Name"]
```

```python
from scipy.optimize import linprog
```

```python
num=len(points)
```

```python
c=-np.array(points) #objective func
```

```python
from sklearn.preprocessing import LabelEncoder

clubs_le=LabelEncoder()
clubs_le.fit(optimisation_df['Team'])
clubs=clubs_le.transform(optimisation_df['Team'])
```

```python
A_ub=[]
A_ub.append(np.array(cost))
for club_id in np.unique(clubs):
    A_ub.append(np.array([1 if clubs[i]==club_id else 0 for i in range(num)]))

A_ub=np.array(A_ub)

goalkeepers=np.array([1 if position[i]=='GKP' else 0 for i in range(num)])
defenders=np.array([1 if position[i]=='DEF' else 0 for i in range(num)])
midfielders=np.array([1 if position[i]=='MID' else 0 for i in range(num)])
forwards=np.array([1 if position[i]=='FWD' else 0 for i in range(num)])

A_eq=np.array([np.ones(num), goalkeepers, defenders, midfielders, forwards])
b_eq=[15, 2, 5, 5, 3]
```

```python
defenders
```

```python
np.ones(num)
```

```python
A_ub
```

```python
A_ub.shape
```

```python
integrality=np.full(num, 3)
integrality
```

```python
result=linprog(
    c=c,
    A_ub=A_ub,
    b_ub=[100]+[3]*20,
    A_eq=A_eq,
    b_eq=[15, 2, 5, 5, 3],
    integrality=integrality,
    bounds=(0,1)
)
```

```python
clubs=clubs_le.inverse_transform(clubs)
clubs
```

```python
result.x
```

```python
total_cost=0
total_points=0
if result.success:
    for i in range(len(result.x)):
        if result.x[i]!=0:
            total_cost+=cost[i]
            total_points+=points[i]
            print(f"{name[i]} - {clubs[i]} - {position[i]}; Points: {points[i]} Price: {cost[i]}")
```
